```python
import yfinance as yf
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the ticker symbols for each coin
tickers = ["BTC-USD", "ETH-USD", "BNB-USD", "SOL-USD", "USDT-USD"]

# Fetch historical price data for each coin
coin_data = {}
for ticker in tickers:
    coin_ticker = yf.Ticker(ticker)
    if f"{ticker.lower()}.csv" in os.listdir():
        coin_data[ticker] = pd.read_csv(f"{ticker.lower()}.csv", index_col=0)
    else:
        coin_data[ticker] = coin_ticker.history(period="max")
        coin_data[ticker].to_csv(f"{ticker.lower()}.csv")

```

```python
coin_data = {}
for ticker in tickers:
    coin_ticker = yf.Ticker(ticker)
    coin_data[ticker] = coin_ticker.history(period="10y")

# Plot the closing prices of all coins on the same graph
plt.figure(figsize=(10, 6))
for ticker, data in coin_data.items():
    plt.plot(data.index, data['Close'], label=ticker)

plt.title('Closing Prices of Cryptocurrencies (Last 10 Years)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
coin_data = {}
for ticker in tickers:
    coin_ticker = yf.Ticker(ticker)
    coin_data[ticker] = coin_ticker.history(period="7d")

# Plot the closing prices of all coins on the same graph
plt.figure(figsize=(10, 6))
for ticker, data in coin_data.items():
    plt.plot(data.index, data['Close'], marker='o', label=ticker)

plt.title('Closing Prices of Cryptocurrencies (Last 7 Days)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
